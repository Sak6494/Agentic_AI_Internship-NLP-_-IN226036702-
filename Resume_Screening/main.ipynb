```python
job_description = """
Data Scientist Role
Skills required: Python, Machine Learning, SQL, NLP
Experience required: 2 years
"""
```

```python
strong_resume = """
Python developer with 3 years experience in Machine Learning, SQL, NLP.
Worked on data pipelines and predictive modeling.
"""

average_resume = """
Python developer with 1 year experience.
Knows SQL and basic Machine Learning.
"""

weak_resume = """
Fresher with basic Excel and communication skills.
"""
```

```python
resumes = {
    "Strong": strong_resume,
    "Average": average_resume,
    "Weak": weak_resume
}

for name, resume in resumes.items():
    print("\n" + "="*60)
    print(f"CANDIDATE TYPE: {name}")
    print("="*60)

    result = run_pipeline(
        resume_text=resume,
        job_description=job_description,
        tag=name.lower()
    )

    print("\nFINAL OUTPUT:\n")
    print(result)
    
```
