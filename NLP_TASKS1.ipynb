#TASK 1:Conceptual Understanding 

1) What is the difference between "Love" and "love" in NLP?
Ans-> In NLP, the difference between "Love" and "love" is because of capital letters. A computer can treat them as two different words since one starts with a capital letter and the other does not. "Love" may appear at the beginning of a sentence or be used for emphasis, while "love" is the normal form. But in most NLP tasks, we convert all words into lowercase, so both "Love" and "love" are treated as the same word. This makes the data simpler and avoids confusion for the model

2) What happens if stopwords are not removed?
Ans-> Stopwords are very common words like "the", "is", "and", "in", etc. If we don’t remove them, the text becomes a bit messy for the model. These words appear everywhere but don’t add much meaning, so they can distract the model from focusing on the important words. This can also make processing slower.

3) Two real-world scenarios where removing stopwords can be harmful.
Ans-> Removing stopwords can sometimes completely change the meaning of a sentence. For example, in sentiment analysis, the sentence "I do not like this movie" becomes "like movie" if we remove stopwords. Now the meaning looks positive instead of negative, which is wrong. Another example is in search engines or question-answering systems. If someone asks "What is the best way to learn NLP?", removing stopwords gives "best way learn NLP". This removes the question feeling and makes it harder for the system to understand what the user actually wants.

4)  Difference between stemming and lemmatization
-> Both stemming and lemmatization are used to simplify words, but they do it differently. Stemming is like cutting a word quickly without thinking too much—it just removes endings. For example, "studies" might become "studi", which isn’t even a proper word. Lemmatization, on the other hand, is smarter. It understands the meaning and grammar of the word and converts it into the correct base form, like "studies" to "study" or "better" to "good".

```python
#TASK2:Build Advanced Preprocessing Function


import re

def preprocess_text(text):
    
    # Handle invalid input
    if not text or not isinstance(text, str):
        return [], ""
    
    # Lowercase
    text = text.lower()
    
    # Remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)
    
    # Remove emails
    text = re.sub(r'\S+@\S+', '', text)
    
    # Remove numbers
    text = re.sub(r'\d+', '', text)
    
    # Handle repeated characters
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    
    # Remove special characters
    text = re.sub(r'[^a-z\s]', '', text)
    
    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()
    
    # Handle empty after cleaning
    if not text:
        return [], ""
    
    # Remove short words (≤2), keep 'no', 'not'
    words = text.split()
    words = [w for w in words if len(w) > 2 or w in ['no', 'not']]
    
    return words, " ".join(words)
```

```python
#TASK3:Stress Testing
texts = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

for text in texts:
    tokens, clean = preprocess_text(text)
    
    print("\nOriginal Text:", text)
    print("Cleaned Tokens:", tokens)
    print("Cleaned Sentence:", clean)
```

```python
#TASK4:Token Analytics

for text in texts:
    tokens, clean = preprocess_text(text)
    
    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_length = sum(len(w) for w in tokens) / total_tokens if total_tokens else 0
    
    print("\nOriginal Text:", text)
    print("Total Tokens:", total_tokens)
    print("Unique Tokens:", unique_tokens)
    print("Average Token Length:", round(avg_length, 2))
```

```python
#TASK 5: Frequency Analysis
from collections import Counter

all_tokens = []

for text in texts:
    tokens, _ = preprocess_text(text)
    all_tokens.extend(tokens)

counter = Counter(all_tokens)

print("Top 10 most frequent words:")
print(counter.most_common(10))

print("\nTop 5 least frequent words:")
print(counter.most_common()[-5:])
```

```python
#TASK 6: Full Pipeline
def full_pipeline(text_list):
    
    all_tokens = []
    clean_sentences = []
    
    for text in text_list:
        tokens, clean = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(clean)
    
    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }


# Example
result = full_pipeline(texts)
print(result)
```
